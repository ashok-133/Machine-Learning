```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data_clustering1.csv')
```

```python
x1 = data.x1
x2 = data.x2
```

```python
C1 = [[1],[1600]]
C2 = [[4],[400]]
```

```python
def dist_measure(x1,x2,C1,C2):
    dist1 = np.zeros(len(x1))
    dist2 = np.zeros(len(x1))
    for i in range(0,len(x1)):
        dist1[i] = ((x1[i]-C1[0])**2 + (x2[i] - C1[1])**2)**0.5
        dist2[i] = ((x1[i]-C2[0])**2 + (x2[i] - C2[1])**2)**0.5
    return dist1,dist2
def cluster_points(x1,x2,dist1,dist2):
    k = 0
    p = 0
    cluster_1 = np.zeros((len(x1),2))
    cluster_2 = np.zeros((len(x2),2))
    for i in range(0,len(x1)):
        if(dist1[i]<dist2[i]):
            cluster_1[k,0] = x1[i]
            cluster_1[k,1] = x2[i]
            k = k+1
        else:
            cluster_2[p,0] = x1[i]
            cluster_2[p,1] = x2[i]
            p = p+1
    cluster_1 = cluster_1[0:k,:]
    cluster_2 = cluster_2[0:p,:]
    return cluster_1,cluster_2
```

```python
dist1,dist2 = dist_measure(x1,x2,C1,C2)
```

```python
cluster_1,cluster_2 = cluster_points(x1,x2,dist1,dist2)
```

```python
import statistics as st
```

```python
def calc_mean(cluster_1,cluster_2):
    mean_00 = st.mean(cluster_1[:,0])
    mean_01 = st.mean(cluster_1[:,1])
    mean_10 = st.mean(cluster_2[:,0])
    mean_11 = st.mean(cluster_2[:,1])
    return mean_00,mean_01,mean_10,mean_11
```

```python
calc_mean(cluster_1,cluster_2)
```

```python
def k_mean_cluster(x1,x2,C1,C2,iter_max):
    n = 0
    for i in range(0,iter_max):
        dist1,dist2 = dist_measure(x1,x2,C1,C2)
        print(dist_measure(x1,x2,C1,C2))
        cluster_1,cluster_2 = cluster_points(x1,x2,dist1,dist2)
        C1[0],C1[1],C2[0],C2[1] = calc_mean(cluster_1,cluster_2)
        n = n+1;
        print(calc_mean(cluster_1,cluster_2))
    return C1,C2,n
        
```
